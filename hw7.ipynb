```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import itertools
```

```python
data = pd.read_csv('blobs2.csv').values
x = np.array(data[:, :2])
labels = np.array(data[:, 2])
```

```python
x_train, x_eval, labels_train, labels_eval = train_test_split(x, labels, train_size=0.8)
```

```python
print("Linear")
clf = SVC(gamma='scale', kernel='linear', degree=1, probability=True)
clf = clf.fit(x_train, labels_train)
pr = clf.predict(x_eval)
acc = len(pr[pr == labels_eval]) / len(pr)
print("Accuracy " + str(acc))
for i in range(len(x_train)):
    if labels_train[i] == 1:
        plt.scatter(x_train[i][0], x_train[i][1], c='g')
    else:
        plt.scatter(x_train[i][0], x_train[i][1], c='b')
x = np.linspace(np.min(x_eval[:, 0]), np.max(x_eval[:, 0]), 1000)
y = np.linspace(np.min(x_eval[:, 1]), np.max(x_eval[:, 1]), 1000)
grid = np.array([x for x in itertools.product(x, y)])
z = clf.predict_proba(grid)[:, 0]
pg = grid[np.abs(z - 0.5) < 0.01]
plt.plot(pg[:, 0], pg[:, 1], c='r')
```
